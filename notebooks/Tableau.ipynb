```python
def calculate_eps(net_income, weighted_average_shs_out):
    # Use the .replace method to replace 0 with NaN to avoid division by zero
    weighted_average_shs_out_replaced = weighted_average_shs_out.replace(0, np.nan)
    # Now perform element-wise division
    eps = net_income / weighted_average_shs_out_replaced
    return eps

# Call the function
eps = calculate_eps(df['netIncome'], df['weightedAverageShsOut'])
df['eps'] = eps
```

```python
import pandas as pd
import numpy as np  # Import numpy for handling NaN values

# Assuming df is your DataFrame
# Replace zeros with NaN to avoid division by zero errors
df['revenue'] = df['revenue'].replace(0, np.nan)
# Calculate profit margin
df['profit_margin'] = (df['netIncome'] / df['revenue']) * 100
```

```python
import pandas as pd
import numpy as np  # Import numpy to handle NaN values

# Assuming df is your DataFrame
# Replace zeros with NaN to avoid division by zero errors
df['weightedAverageShsOut'] = df['weightedAverageShsOut'].replace(0, np.nan)
df['eps'] = df['eps'].replace(0, np.nan)  # Assuming eps is a column in your DataFrame

# Perform vectorized operations to calculate the Market Value Per Share and PE Ratio
df['market_value_per_share'] = df['marketCapitalization'] / df['weightedAverageShsOut']
df['pe_ratio'] = df['market_value_per_share'] / df['eps']
```
